```python
nba = pd.read_excel('nba.xlsx')
nba
```

```python
nba.shape
```

```python
nba.describe()
```

```python
nba.isna().sum()
```

```python
nba['Salary'].fillna(nba['Salary'].mean(), inplace=True)
```

```python
nba.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
nba_new = pd.DataFrame(imputer.fit_transform(nba),columns=nba.columns,index=nba.index)
nba_new = nba_new.astype(nba.dtypes)
nba = nba_new
```

```python
nba.isnull().sum()
```

```python
nba
```

```python
nba.dtypes
```

```python
nba['Weight'] = nba['Weight'].astype(float)
```

```python
nba.duplicated().sum()
```

```python
nba.drop('Height', axis=1, inplace=True)
```

```python
nba.describe()
```

```python
nba['Age Group'] = pd.cut(nba['Age'], bins=[15,20,25,30,35,40,45], include_lowest=True, labels=['15-20','21-25','26-30','31-35','36-40','41-45'])
```
