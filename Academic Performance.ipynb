```python
import numpy as np
import pandas as pd
```

```python
acad = pd.read_csv('Academic_Performance.csv')
acad
```

```python
acad.isnull().sum()
```

```python
cols = ['COURSE 1 MARKS','COURSE 2 MARKS','COURSE 3 MARKS','COURSE 4 MARKS','COURSE 5 MARKS']
acad[cols] = acad[cols].fillna(0)
acad
```

```python
acad.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
acad_new = pd.DataFrame(imputer.fit_transform(acad), columns=acad.columns, index=acad.index)
acad_new = acad_new.astype(acad.dtypes)
acad = acad_new
```

```python
acad.isnull().sum()
```

```python
acad.describe()
```

```python
for i in cols:
    acad = acad[(acad[i] >= 0) & (acad[i] <= 100)]
```

```python
acad.describe()
```

```python
def outliers(data):
    outliers=[]
    data = sorted(data)
    q1 = np.percentile(data,25)
    q3 = np.percentile(data,75)
    iqr = q3-q1
    lower = q1 - (1.5*iqr)
    upper = q3 + (1.5*iqr)
    for i in data:
        if (i<lower or i>upper):
            outliers.append(i)
    print(outliers)
    return lower,upper
```

```python
lower,upper = outliers(acad['COURSE 1 MARKS'])
acad = acad[acad['COURSE 1 MARKS'] > lower]
acad = acad[acad['COURSE 1 MARKS'] < upper]
```

```python
acad.shape
```

```python
performance = acad.select_dtypes(exclude=np.number)
performance
```

```python
performance['PLACEMENT'].unique()
```

```python
performance['PLACEMENT'].replace({'Yes':1, 'No':2},inplace=True)
performance
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
col=['STUDENT_ID','EDUCATION_TYPE','ACADEMIC_PROGRAM','OVEARLL_GRADE']
for i in col:
    performance[i] = label_encoder.fit_transform(performance[i])
```
