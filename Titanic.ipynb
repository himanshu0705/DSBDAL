```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('titanic.csv')
titanic
```

```python
titanic.describe()
```

```python
titanic.isnull().sum()
```

```python
titanic.dtypes
```

```python
titanic['Embarked'].fillna('S',inplace=True)
```

```python
titanic['Cabin'] = np.where(titanic['Cabin'].isnull(),0,1)
```

```python
titanic
```

```python
mean_age = round(titanic['Age'].mean())
std_age = round(titanic['Age'].std())
```

```python
mean_age, std_age
```

```python
lower = mean_age - std_age
upper = mean_age + std_age
```

```python
titanic['Age'].fillna(np.random.randint(lower,upper),inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic
```

```python
sns.countplot(titanic, x='Pclass', hue='Survived')
plt.show()
```

```python
titanic['Age'].unique()
```

```python
titanic['Age'] = titanic['Age'].apply(lambda x: round(x))
```

```python
titanic['Age'].unique()
```

```python
titanic = titanic[titanic['Age'] > 0]
```

```python
sns.histplot(titanic, x='Age', hue='Survived')
plt.show()
```

```python
plt.pie(titanic.groupby('Survived')['PassengerId'].count(), labels=['Not Survived','Survived'],autopct='%1.0f%%')
plt.title('Survival')
plt.show()
```
