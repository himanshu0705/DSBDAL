```python
alexa['cleaned text'] = alexa['verified_reviews'].apply(lambda x: x.lower())
```

```python
import string
string.punctuation
```

```python
alexa['cleaned text'] = alexa['cleaned text'].apply(lambda x: "".join([i for i in x if i not in string.punctuation]))
```

```python
alexa
```

```python
import re
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"
                               u"\U0001F300-\U0001F5FF"
                               u"\U0001F680-\U0001F6FF"
                               u"\U0001F1E0-\U0001F1FF"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+",flags=re.UNICODE)
    return (emoji_pattern.sub(r'',string))
```

```python
alexa['cleaned text'] = alexa['cleaned text'].apply(lambda x: remove_emoji(x))
```

```python
alexa
```

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
# needed for word tokenizer
nltk.download('punkt')
```

```python
alexa['tokenized'] = alexa['cleaned text'].apply(lambda x: word_tokenize(x))
```

```python
alexa['tokenized']
```

```python
from nltk.corpus import stopwords
```

```python
# needed for stopwords
nltk.download('stopwords')
```

```python
sw = set(stopwords.words('english'))
```

```python
alexa['tokenized'] = alexa['tokenized'].apply(lambda x: " ".join([i for i in x if i not in sw]))
```

```python
alexa['tokenized'] = alexa['tokenized'].apply(lambda x: x.split())
```

```python
alexa
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
alexa['stemmed'] = alexa['tokenized'].apply(lambda x: " ".join([ps.stem(word) for word in x]))
```

```python
alexa
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# required for lemmatizer
nltk.download('wordnet')
```

```python
nltk.download('omw-1.4')
```

```python
alexa['lemmatized'] = alexa['tokenized'].apply(lambda x: " ".join([lemmatizer.lemmatize(word) for word in x]))
```

```python
alexa
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
```

```python
x = vectorizer.fit_transform(alexa['lemmatized'])
bow = pd.DataFrame(x.toarray(), columns=vectorizer.get_feature_names_out())
bow
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```python
y = tfidf.fit_transform(alexa['lemmatized'])
df_tfidf = pd.DataFrame(y.toarray(), columns=tfidf.get_feature_names_out())
df_tfidf
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
si = SentimentIntensityAnalyzer()
```

```python
alexa['sentiment score'] = alexa['cleaned text'].apply(lambda x: si.polarity_scores(x)['compound'])
```

```python
def sentiment_func(n):
    if n>0:
        return 'positive'
    elif n<0:
        return 'negative'
    else:
        return 'neutral'
```

```python
alexa['sentiment'] = alexa['sentiment score'].apply(lambda x:sentiment_func(x))
```
