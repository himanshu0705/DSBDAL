```python
import numpy as np
import pandas as pd
```

```python
age_income = pd.read_excel('Age-Income-Dataset.xlsx')
age_income
```

```python
mean = sum(age_income['Income']/len(age_income['Income']))
mean
```

```python
n = len(age_income['Income'])
if n%2:
    median = sorted(age_income['Income'])[round(0.5*(n-1))]
else:
    order,index = sorted(age_income['Income']), round(0.5*n)
    median = 0.5*(order[index-1] + order[index])
median
```

```python
group = age_income.groupby('Income').size().to_dict()
mode = [i for i,j in group.items() if j==max(group.values())]
mode
```
